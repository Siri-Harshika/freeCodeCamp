### Defining a function

```python
# Defining a function f which takes x as input
def f(x):
    y = 4 * x**2 + 3 * x - 2
    return y

# Headings
print("x \t y")

# Loop
for x in range(11):
    print(x, "\t", f(x))
```

### Graph and table of (x, y) values

```python
# Syntax to import graphing library
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

# Setting the dimensions of the graph
xmin = -10
xmax = 10
ymin = -10
ymax = 10

# Defining the dimensions of the graph
plt.axis([xmin, xmax, ymin, ymax])

# Plotting the axis lines
plt.plot([xmin, xmax], [0, 0], 'y') # X-axis
plt.plot([0, 0], [ymin, ymax], 'b') # Y-axis

# Table and Graph of (x, y) values
print("x \t y")

for x in range(xmin, xmax + 1):
    y = 0.5*x + 1
    print(x, "\t", y)         # (x, y) values in table
    plt.plot([x], [y], 'ro')  # (x, y) values in graph

# Showing the graph
plt.show()
```

### Using array as inputs

```python
# Syntax to import graphing library
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

# To use arrays
import numpy as np

# Setting the dimensions of the graph
xmin = -10
xmax = 10
ymin = -10
ymax = 10

# Defining the dimensions of the graph
plt.axis([xmin, xmax, ymin, ymax])

# Drawing the axes
plt.plot([xmin, xmax], [0, 0], 'r') # x axis
plt.plot([0, 0], [ymin, ymax], 'b') # y axis

# Line looks smooth withot much processing time
points = 2*(xmax - xmin)

# Selecting x-coordinates (in an array) which are of even spacing on x-axis
x = np.linspace(xmin, xmax, points)

# Value of y-coordinate (in an array because the values of x are in a array)
y = 3*x + 2

# Plotting the line
plt.plot(x, y, 'pink') # No need for [] because x and y itself represent an array

# Showing the graph
plt.show()
```

### Customizing a graph

```python
# Syntax to import graphing library
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

# To use array
import numpy as np

# Setting and defining the dimensions of the graph
xmin = -10
xmax = 10
ymin = -10
ymax = 10

plt.axis([xmin, xmax, ymin, ymax])

# Plotting and customizing the axes
plt.plot([xmin, xmax], [0, 0], 'k')      # x axis,k for black
ax.set_xticks(np.arange(xmin, xmax, 1)) # Marking points on x axis
ax.set_xlabel('x values')                # Naming the x axis

plt.plot([0, 0], [ymin, ymax], 'brown')  # y axis
ax.set_yticks(np.arange(ymin, ymax, 1)) # Marking points on y axis
ax.set_ylabel('y values')                # Naming the y axis

ax.set_title('Basic graph')  # Naming the graph
ax.grid(True)                # Setting the gridlines

# Line looks smooth without much processing time
points = 2*(xmax - xmin)

# Selecting x-coordinates (in an array) which are of even spacing on x axis
x = np.linspace(xmin, xmax, points)

# Drawing first line
y = 2*x + 3
plt.plot(x, y, 'orange', label = "y = 2x + 3")

# Plotting a point
plt.plot([4], [5], 'ro', label = 'point')

# Drawing another line
plt.plot(x, 3*x, 'grey', label = "steeper line")

# Showing the graph and label (legend)
plt.legend()
plt.show()
```
